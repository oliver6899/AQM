# H1

### H₀: Shots on target per 90 minutes have no correlation with goal contributions per 90 minutes.
### H₁: Players with a higher number of shots on target per 90 minutes have a higher goal contribution per 90 minutes.


```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset (tilpas stien hvis nødvendig)
data = pd.read_csv("https://raw.githubusercontent.com/aaubs/AQM-2025/refs/heads/main/EPL-Players.csv")

# Quick preview
data.head()

```

```python
import numpy as np

# Vi tager kun spillere med minutter > 0 for at undgå division med 0
data = data[data['Minutes'] > 0].copy()

# Shots on target per 90
data['sot90'] = data['Shots On Target'] / (data['Minutes'] / 90)

# Goal contributions = goals + assists
data['gc'] = data['Goals'] + data['Assists']

# Goal contributions per 90
data['gc90'] = data['gc'] / (data['Minutes'] / 90)

```

# H2

### H₀: Passing accuracy (%) has no correlation with goal contributions per 90 minutes.
### H₁: Higher passing accuracy (%) is positively correlated with goal contributions per 90 minutes.





```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Vi fortsætter i samme 'data' som i H1
# Hvis ikke, så sørg for at have læst EPL-Players.csv ind først.

# Kun spillere med minutter > 0 og gyldig Passes%
df_h2 = data[(data['Minutes'] > 0) & data['Passes%'].notna()].copy()

# Tjek de to centrale variabler til H2
df_h2[['Passes%', 'gc90']].info()
df_h2[['Passes%', 'gc90']].describe()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x=df_h2['Passes%'], y=df_h2['gc90'], ci=None)
plt.xlabel("Passing accuracy (Passes%)")
plt.ylabel("Goal contributions per 90 (gc90)")
plt.title("Relationship between Passing Accuracy and Goal Contributions per 90")
plt.grid(True)
plt.show()
```

```python
# Definér X og y til H2
X_h2 = sm.add_constant(df_h2['Passes%'])  # konstant + Passes%
y_h2 = df_h2['gc90']

# Estimér modellen
model_h2 = sm.OLS(y_h2, X_h2).fit()

# Få hele regressions-tabellen
print(model_h2.summary())
```

```python
residuals_h2 = model_h2.resid
fitted_h2 = model_h2.fittedvalues

plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_h2, y=residuals_h2)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted values (predicted gc90)")
plt.ylabel("Residuals")
plt.title("H2 – Residual Plot")
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(residuals_h2, kde=True)
plt.title("H2 – Residual Distribution")

# Q-Q plot
plt.subplot(1, 2, 2)
stats.probplot(residuals_h2, dist="norm", plot=plt)
plt.title("H2 – Q-Q Plot")

plt.show()
```

```python
influence_h2 = model_h2.get_influence()
std_resid_h2 = influence_h2.resid_studentized_internal

# Outliers: observationer med |studentized residual| > 3
outliers_mask_h2 = np.abs(std_resid_h2) > 3
outliers_h2 = df_h2[outliers_mask_h2]

outliers_h2
```

```python
# Behold kun observationer uden ekstreme outliers
df_h2_clean = df_h2[~outliers_mask_h2].copy()

X_h2_clean = sm.add_constant(df_h2_clean['Passes%'])
y_h2_clean = df_h2_clean['gc90']

model_h2_clean = sm.OLS(y_h2_clean, X_h2_clean).fit()
print(model_h2_clean.summary())
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(
    x=df_h2_clean['Passes%'],
    y=df_h2_clean['gc90'],
    ci=None
)

plt.xlabel("Passing accuracy (Passes%)")
plt.ylabel("Goal contributions per 90 (gc90)")
plt.title("Passing Accuracy vs GC per 90 (Outliers Removed)")
plt.grid(True)
plt.show()

```
